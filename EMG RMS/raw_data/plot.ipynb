```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
df = pd.read_excel("P14.xlsx")

# Remove duplicated "Time(s)" column
df = df.loc[:,~df.columns.duplicated()]

# Remove duplicated "Time(s)" column
df = df.loc[:,~df.columns.duplicated()]

# Drop duplicated "X[s]" columns except the first one
cols_to_keep = [col for col in df.columns if 'X[s]' not in col or df.columns.get_loc(col) == 0]
df = df[cols_to_keep]

# Save to new CSV file
df.to_csv("p14-plot.csv", index=False)
```

```python
# Read the newly created CSV file
df_plot = pd.read_csv("p14-plot.csv")

# Plot the data from the new CSV file
plt.figure(figsize=(12, 6))  # Adjust the figure size
for column in df_plot.columns[1:]:
    plt.plot(df_plot[df_plot.columns[0]], df_plot[column], label=column, linestyle='-')

plt.xlabel(df_plot.columns[0])
plt.ylabel("Value")
plt.title("Plot of Data from p14-plot.csv")
plt.legend()
plt.grid(True)
plt.tight_layout()  # Increase gap between time points
plt.show()
```
